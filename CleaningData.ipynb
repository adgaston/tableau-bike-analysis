```python
import glob
import os
import pandas as pd
```

```python
path = '/Users/amandagaston/Desktop/CLASS/02-Homework/20-Tableau/surlatableau/Resources/TripHistories/FullData'
```

```python
csvfiles = glob.glob(os.path.join(path, '*.csv'))
```

```python
bikedata = []
for csvfile in csvfiles:
    record = pd.read_csv(csvfile)
    bikedata.append(record)
```

```python
fulldata = pd.concat(bikedata, ignore_index=True)
```
